## Q1. What is the mathematical formula for a linear SVM?

A linear Support Vector Machine (SVM) is a binary classification algorithm used to find a hyperplane that best separates two classes in a feature space. The mathematical formula for a linear SVM can be represented as follows:

Given a dataset with N samples and D features, the input data can be represented as X = {x1, x2, ..., xN}, where xi is a D-dimensional feature vector. The corresponding binary labels for the data are represented as Y = {y1, y2, ..., yN}, where yi is either +1 or -1, indicating the class of each sample.

The goal of the linear SVM is to find a hyperplane defined by the equation:

w^T * x + b = 0

where:

w is the weight vector perpendicular to the hyperplane (also known as the normal vector).
x is the input feature vector.
b is the bias term, representing the offset of the hyperplane from the origin.
The decision function for predicting the class of a new sample x_new is given by:

f(x_new) = sign(w^T * x_new + b)

where sign is the sign function, returning +1 if the argument is positive or zero, and -1 if the argument is negative.

The goal of the SVM training process is to find the optimal values for w and b that maximize the margin between the two classes while minimizing classification errors. This is achieved by solving the following optimization problem:

Minimize: ||w||^2 + C * Σ(max(0, 1 - y_i * (w^T * x_i + b)))

subject to: y_i * (w^T * x_i + b) ≥ 1 for all training samples (x_i, y_i)

where:

||w||^2 represents the L2 norm of the weight vector w.
C is the regularization parameter that controls the trade-off between maximizing the margin and minimizing the classification errors. A higher C value leads to a smaller margin but fewer misclassifications on the training data, while a lower C value results in a larger margin but may allow more misclassifications.
The optimization problem is typically solved using techniques such as the quadratic programming solver to find the optimal values of w and b, which define the hyperplane that best separates the two classes.

## Q2. What is the objective function of a linear SVM?

The objective function of a linear Support Vector Machine (SVM) is the mathematical expression that the SVM aims to minimize during the training process. The objective function consists of two main components: the margin maximization term and the regularization term.

Margin Maximization Term:
The SVM aims to find a hyperplane that maximizes the margin between the two classes in the feature space. The margin is the distance between the hyperplane and the closest data points from each class. The larger the margin, the more confident the SVM is about its classification decisions.
The margin can be computed as the distance between two parallel hyperplanes, one for each class. Let's assume we have two hyperplanes: one for the positive class (+1) and the other for the negative class (-1). These hyperplanes are defined by the equations:

w^T * x + b = 1 (for the positive class)
w^T * x + b = -1 (for the negative class)

The margin (M) between these two hyperplanes is given by:

M = (2 / ||w||)

where ||w|| is the L2 norm (Euclidean norm) of the weight vector w.

To maximize the margin, we need to minimize ||w||. Thus, the margin maximization term of the objective function is:

Minimize: ||w||^2

Regularization Term:
The SVM also introduces a regularization term to control the trade-off between maximizing the margin and minimizing classification errors. The regularization term prevents the SVM from overfitting the training data. It is represented as the sum of the classification errors for the training samples.
For each training sample (x_i, y_i), where x_i is the feature vector and y_i is the corresponding binary label (+1 or -1), the classification error is given by:

ξ_i = max(0, 1 - y_i * (w^T * x_i + b))

The regularization term penalizes the SVM for making classification errors by adding up all the ξ_i values for misclassified samples.

The final objective function of a linear SVM is the combination of the margin maximization term and the regularization term, controlled by the regularization parameter C:

Minimize: ||w||^2 + C * Σ(max(0, 1 - y_i * (w^T * x_i + b)))

subject to: y_i * (w^T * x_i + b) ≥ 1 for all training samples (x_i, y_i)

The optimization problem is to find the optimal values of w and b that minimize this objective function, resulting in the hyperplane that best separates the two classes while balancing the margin and classification errors according to the value of C.

## Q3. What is the kernel trick in SVM?

The kernel trick is a fundamental concept in Support Vector Machines (SVM) that allows SVMs to efficiently handle non-linearly separable data. SVMs are originally designed for binary classification and work well when the data is linearly separable. However, many real-world datasets are not linearly separable, and in such cases, the kernel trick comes to the rescue.

The basic idea behind the kernel trick is to implicitly transform the original feature space into a higher-dimensional space where the data becomes linearly separable. This transformation is achieved by using a kernel function, which is a similarity measure between two data points in the original feature space.

Mathematically, let's consider a training dataset X = {x1, x2, ..., xn}, where xi is a D-dimensional feature vector, and the corresponding binary labels Y = {y1, y2, ..., yn}, where yi is either +1 or -1, indicating the class of each sample.

In the original feature space, the decision function for a linear SVM can be written as:

f(x) = sign(w^T * x + b)

where:

w is the weight vector perpendicular to the hyperplane.
x is the input feature vector.
b is the bias term.
To introduce non-linearity, the kernel trick replaces the dot product (inner product) w^T * x with a kernel function K(x, x'), which calculates the similarity between two data points x and x' in the original feature space. The decision function then becomes:

f(x) = sign(Σ(α_i * y_i * K(x, x_i)) + b)

where:

α_i are the coefficients (Lagrange multipliers) obtained during the training process.
K(x, x_i) is the kernel function that measures the similarity between x and each support vector x_i.
The key advantage of using the kernel trick is that it allows the SVM to operate in the higher-dimensional transformed space without explicitly computing the transformation. The kernel function effectively computes the dot product in the higher-dimensional space, avoiding the need to explicitly represent the transformed data points.

Some commonly used kernel functions are:

Linear Kernel: K(x, x') = x^T * x'
Polynomial Kernel: K(x, x') = (γ * x^T * x' + r)^d
Gaussian (RBF) Kernel: K(x, x') = exp(-γ * ||x - x'||^2)
The choice of the kernel function and its hyperparameters (e.g., γ, d) depends on the data and the specific problem at hand. By using appropriate kernel functions, SVMs can effectively handle non-linearly separable data and provide powerful classification capabilities

## Q4. What is the role of support vectors in SVM Explain with example

In Support Vector Machines (SVM), support vectors play a crucial role in determining the optimal hyperplane that separates the two classes in a binary classification problem. Support vectors are the data points from the training set that lie closest to the separating hyperplane or, in some cases, lie on the wrong side of the hyperplane. These points are the most informative and influential in defining the decision boundary, and they "support" the construction of the classifier.

Let's explain the role of support vectors with a simple example:

Example:
Consider a binary classification problem with two classes, represented by the following data points in a 2-dimensional feature space:

Class 1 (Positive class):
A: (2, 3)

B: (4, 2)

C: (3, 4)


Class 2 (Negative class):


D: (6, 5)

E: (7, 6)

F: (8, 7)


We want to find a linear SVM to separate these two classes.

Step 1: Train the SVM
The SVM algorithm aims to find the hyperplane that maximizes the margin between the two classes. To do this, it selects a subset of the training data points as support vectors. In this example, let's say that points A, B, D, and E are chosen as support vectors because they lie either on the margin boundary or on the wrong side of the hyperplane.

Step 2: Find the Optimal Hyperplane
The SVM computes the optimal hyperplane based on the support vectors. The hyperplane is represented as w^T * x + b = 0, where w is the weight vector perpendicular to the hyperplane, x is the input feature vector, and b is the bias term.

Step 3: Make Predictions
Once the optimal hyperplane is determined, the SVM can classify new data points. For any new input data point x_new, the SVM uses the decision function:

f(x_new) = sign(w^T * x_new + b)

where sign is the sign function returning +1 for positive class and -1 for the negative class.

In this example, the support vectors A, B, D, and E are crucial in defining the decision boundary. They determine the orientation and position of the hyperplane that separates the two classes. Other data points in the training set do not contribute to defining the hyperplane because they are not as critical to the decision boundary.

The reason SVM focuses on the support vectors is to ensure robustness and better generalization. By considering only the support vectors, SVM avoids being influenced by irrelevant or redundant data points, which can lead to overfitting.

In summary, support vectors are the key elements in SVM that determine the decision boundary and play a critical role in achieving an optimal classification model that generalizes well to new, unseen data.

## Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

To illustrate the concepts of hyperplane, marginal plane, soft margin, and hard margin in SVM, let's consider a simple 2-dimensional dataset with two classes. We will visualize the data points and the decision boundaries using graphs.

Example Data:

Class 1 (Positive class):

(2, 3), (3, 3), (3, 4), (4, 4)


Class 2 (Negative class):

(1, 1), (1, 2), (2, 2), (2, 1)


Let's create the graphs to illustrate each concept:


**Hyperplane:**

In a linear SVM, the hyperplane is the decision boundary that separates the two classes. In 2D, the hyperplane is a line. In higher dimensions, it becomes a hyperplane.
Graph:
The black line represents the hyperplane that separates the two classes.

**Marginal Plane:**

The marginal plane, also known as the margin, is the region around the hyperplane that contains the support vectors. The distance between the margin and the hyperplane is called the margin width. In a hard margin SVM, all support vectors lie on the margin boundary, and the margin width is maximized.
Graph:
The shaded regions around the hyperplane represent the marginal planes. The support vectors are marked with larger data points.

**Hard Margin:**

In a hard margin SVM, the margin width is maximized, and the SVM aims to find a hyperplane that perfectly separates the two classes with no data points allowed within the margin.
Graph:
The margin width is maximized, and the SVM correctly separates the two classes without any misclassifications.

**Soft Margin:**

In real-world datasets, it is common to have overlapping data points or noise that makes it difficult to find a perfect separation. Soft margin SVM allows some misclassifications to find a hyperplane with a smaller margin that still separates the classes while minimizing the classification errors.

**Graph:**

The soft margin SVM allows some misclassifications (represented by the dashed lines) to accommodate the overlapping data points. The margin width is reduced to find a more flexible decision boundary.

## Q6. SVM Implementation through Iris dataset.

###  Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
X = dataset.data
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y ,train_size=0.30,random_state=42)
```

### Train a linear SVM classifier on the training set and predict the labels for the testing setl

```python
from sklearn.svm import SVC
svc = SVC(kernel="linear")
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

## Compute the accuracy of the model on the testing setl

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
